```python
#Ajout de fonction de netoyage 
def clean_data(commentaire):
    commentaire=''.join(commentaire)
    commentaire = commentaire.translate(str.maketrans('', '', string.punctuation))
    
    # 3. Supprimer les chiffres
    commentaire = re.sub(r'\d+', '', commentaire)
    
    # 4. Tokenisation simple (split par espace)
    commentaire = commentaire.split()
    return commentaire
```

```python
import string
import spacy

#Tokenisation
nlp=spacy.load("fr_core_news_sm")

#Deja on va supprimer les poctuation
#Transformation de notre liste en string
def pretraitement_tokenisation(file_name):
    with open(file_name,"r",encoding='utf-8') as file:
        commentaire_clean=file.readlines()
        #netoyage
        commentaire_clean=clean_data(commentaire_clean)
        commentaire_clean=' '.join(commentaire_clean)
        commentaire_clean=nlp(commentaire_clean)
    return commentaire_clean
    
```

```python
#Now let's traine the model 3.	Entraîner un modèle de classification Naive Bayes pour prédire la classe d'un texte.
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix

# -------------------
# 1. Créer dataset
# -------------------
# Fusionner les commentaires positifs et négatifs
X = commentaire_positif_clean_lem + commentaire_negatif_clean_lem
y = [1] * len(commentaire_positif_clean_lem) + [0] * len(commentaire_negatif_clean_lem)

# -------------------
# 2. Vectorisation
# -------------------
vectorizer = TfidfVectorizer()
X_vect = vectorizer.fit_transform(X)

# -------------------
# 3. Séparer en train/test
# -------------------
X_train, X_test, y_train, y_test = train_test_split(X_vect, y, test_size=0.2, random_state=42)

# -------------------
# 4. Entraîner Naive Bayes
# -------------------
model = MultinomialNB()
model.fit(X_train, y_train)

# -------------------
# 5. Évaluer le modèle
# -------------------
y_pred = model.predict(X_test)

print("Rapport de classification :")
print(classification_report(y_test, y_pred))

print("Matrice de confusion :")
print(confusion_matrix(y_test, y_pred))

# -------------------
# 6. Tester sur un nouveau commentaire
# -------------------
nouveau_commentaire = ["Messi es nul"]
nouveau_vect = vectorizer.transform(nouveau_commentaire)
prediction = model.predict(nouveau_vect)

print(" Prédiction :", "Positif" if prediction[0]==1 else "Négatif")
```
