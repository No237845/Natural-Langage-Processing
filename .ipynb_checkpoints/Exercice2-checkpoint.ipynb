```python

#Pos Tagging
import spacy
```

```python

nlp = spacy.load("fr_core_news_sm")
```

```python
import fr_core_news_sm
```

```python
nlp = fr_core_news_sm.load()
```

```python
#Debut tokenisation 
with open ("fichier_fusionne.csv","r",encoding='utf-8') as file:
    text=file.readlines()
text=' '.join(text)
document=nlp(text)
for token in document:
    print(token ," | ",token.pos_ ," | ", spacy.explain(token.pos_))
#Fin tokennisation
```

```python
#3.	Effectuer la tokenisation et le POS tagging (Part of Speech) sur le texte avec NLTK.
import nltk
#Installation du package punkt_tab
nltk.download('punkt_tab')
```

```python
#Debut Tokenisation
with open ("fichier_fusionne.csv","r",encoding="utf-8") as file:
    text=file.readlines()
text=' '.join(text)
document=nltk.word_tokenize(text)
for doc in document:
    print(doc)

#fin Tokenisation
```

```python
#Installation du package averaged_perceptron_tagger_eng
nltk.download('averaged_perceptron_tagger_eng')
```

```python
#POS tagging (Part of Speech) sur le texte avec NLTK

tagged=nltk.pos_tag(document)
for tag in tagged:
    print(nltk.pos_tag(tag))
```
