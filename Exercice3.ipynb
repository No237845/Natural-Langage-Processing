```python
import csv
import spacy
import string
```

```python
with open("fichier_fusionne.csv","r",encoding="utf-8") as file:
    text=file.readlines()
text=' '.join(text)
#Suppression de la ponctuation
texte_sans_ponctuation = text.translate(str.maketrans('', '', string.punctuation))
texte_sans_ponctuation=texte_sans_ponctuation.lower()
```

```python
# TÃ©lÃ©chargement des stop words 
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Suppression des stop words
mots = word_tokenize(texte_sans_ponctuation)  # Tokenisation et conversion en minuscules
stop_words = set(stopwords.words('french'))  # Stop words en franÃ§ais
mots_filtrÃ©s = [mot for mot in mots if mot not in stop_words]
mots
mots_filtrÃ©s
```
