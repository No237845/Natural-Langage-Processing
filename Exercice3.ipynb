```python
import csv
import spacy
import string
```

```python
with open("fichier_fusionne.csv","r",encoding="utf-8") as file:
    text=file.readlines()
text=' '.join(text)
#Suppression de la ponctuation
texte_sans_ponctuation = text.translate(str.maketrans('', '', string.punctuation))
texte_sans_ponctuation=texte_sans_ponctuation.lower()
```

```python
# Téléchargement des stop words 
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Suppression des stop words
mots = word_tokenize(texte_sans_ponctuation)  # Tokenisation et conversion en minuscules
stop_words = set(stopwords.words('french'))  # Stop words en français
mots_filtrés = [mot for mot in mots if mot not in stop_words]
mots
mots_filtrés
```

```python
import spacy
#Stemming is the process of producing morphological variants of a root/base word.
#    Stemming programs are commonly referred to as stemming algorithms or stemmers.
#    A stemming algorithm reduces the words “chocolates”, “chocolatey”, and “choco” to the root word, “chocolate” 
#and “retrieval”, “retrieved”, “retrieves” reduce to the stem “retrieve”.
```

```python
#Pour le Stemming
from nltk.stem.snowball import SnowballStemmer
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# Initialisation du stemmer pour le français
#stemmer = SnowballStemmer("french")
ps = PorterStemmer()
mots_stemmes=[]

#for mot in mots_filtrés:
#    mots_stemmes.append(stemmer.stem(mot))
for mot in mots_filtrés:
    mots_stemmes.append(ps.stem(mot))
    
print(mots_stemmes)
```

```python
#Pour la lemmatisation
import spacy

# Charger le modèle français
nlp = spacy.load("fr_core_news_sm")

# Rejoindre les mots filtrés pour reformer une phrase
doc = nlp(" ".join(mots_filtrés))

# Extraire les lemmes
lemmes = [token.lemma_ for token in doc if not token.is_stop]

print("Avant lemmatisation :", mots_filtrés[:20])
print("Après lemmatisation :", lemmes[:20])
```

```python
#Pour la lemmatisation
import spacy

nlp = spacy.load("fr_core_news_sm")
#Conversion de la liste en string
mots_filtrés=' '.join(mots_filtrés)
doc = nlp(mots_filtrés)

for token in doc:
    print(token.text, token.lemma_)
```

```python
L’utilité de la lemmatisation : La lemmatisation est utilisée dans de nombreuses tâches de NLP pour simplifier et normaliser le texte.
    Voici quelques exemples d’utilisation de la lemmatisation :

Modélisation de sujets : En lemmatisant les mots d’un texte, nous pouvons regrouper les différentes formes d’un même mot sous un lemme commun, 
ce qui facilite l’identification des sujets et des thèmes récurrents dans un corpus de documents.
Analyse de sentiments : La lemmatisation permet de réduire les différentes formes lexicales d’un mot à un lemme commun, ce 
qui facilite l’analyse des sentiments associés à ce mot dans un texte. Par exemple, les formes « heureux », « heureuse » et 
« heureusement » peuvent être 
lemmatisées en « heureux », permettant ainsi d’obtenir une représentation plus cohérente des sentiments liés à ce concept.
```
